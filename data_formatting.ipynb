```python
# Load in BLASTP results
blastp_paeni = pd.read_csv('/Users/isabellalin/Library/CloudStorage/Box-Box/Feaga Lab/Cassidy Prince/Bella/Bioinformatics for Paper/BLASTp/psi_blast_paeni_it1.csv', header=None, sep='\t')
blastp_paeni.columns = ['query', 'subject', '%_identity', 'alignment_length', 'mismatches', 'gap_opens', 'q_start', 'q_end', 's_start', 's_end', 'evalue', 'bit_score']

# Drop hits if E-value is >= 0.05
blastp_paeni = blastp_paeni[blastp_paeni['evalue'] < 0.05]

# Make a column for genome assembly and protein accession 
blastp_paeni[['genome', 'protein']] = blastp_paeni['subject'].str.extract(r'^(GCF_[0-9]+\.[0-9]+)_(WP_\S+)$')
```

```python
# Keep the highest bit score for each query-genome pair
blastp_paeni.loc[:, 'bit_score'] = blastp_paeni['bit_score'].astype(float)
blastp_paeni_top = blastp_paeni.sort_values('bit_score', ascending=False).drop_duplicates(subset=['query', 'genome'], keep='first')
blastp_paeni_top = blastp_paeni_top.reset_index(drop=True)
```

```python
# Create a gene presence/absence matrix
blastp_gene_matrix = pd.crosstab(blastp_paeni_top['genome'], blastp_paeni_top['query'])
blastp_gene_matrix.index.name = None
blastp_gene_matrix.columns.name = None
```

```python
# Export
blastp_gene_matrix.to_excel('/Users/isabellalin/Library/CloudStorage/Box-Box/Feaga Lab/Cassidy Prince/Bella/Bioinformatics for Paper/BLASTp/blastp_paeni_gene_matrix.xlsx', index=True)
```

```python
# Import genome info excel
genome_info = pd.read_excel('/Users/isabellalin/Library/CloudStorage/Box-Box/Feaga Lab/Cassidy Prince/Bella/Bioinformatics for Paper/BLASTp/genome_info.xlsx', index_col=0)

# Merge blastp_gene_matrix with genome_info
blastp_gene_matrix_metadata = blastp_gene_matrix.merge(genome_info, left_index=True, right_index=True, how='left')
```

```python
# Export
blastp_gene_matrix_metadata.to_excel('/Users/isabellalin/Library/CloudStorage/Box-Box/Feaga Lab/Cassidy Prince/Bella/Bioinformatics for Paper/BLASTp/blastp_metadata.xlsx', index=True)
```

```python
# Import Spo0B PHMMER output
hmm_results = pd.read_fwf("/Users/isabellalin/Library/CloudStorage/Box-Box/Feaga Lab/Cassidy Prince/Bella/Bioinformatics for Paper/Paeni2_Spo0B_PHMMER_5e2.csv", skiprows = 3, skipfooter = 10, header = None)
column_names = ['target_name', 'target_accession', 'query_name', 'query_accession',
    'full_E-value', 'full_score', 'full_bias',
    'domain_E-value', 'domain_full_score', 'domain_full_bias',
    'exp', 'reg', 'clu', 'ov', 'env', 'dom', 'rep', 'inc', 'description_of_target']

hmm_results.columns = column_names

hmm_results[["genome", "protein"]] = hmm_results["target_name"].str.extract(r"^(GCF_\d+\.\d+)_(\w+_\d+\.\d+)$")
```

```python
# Keep best hit for for each genome
hmm_results = hmm_results.sort_values(by=['genome', 'full_E-value'])
hmm_results = hmm_results.drop_duplicates(subset=['genome'], keep='first')
hmm_results = hmm_results.reset_index(drop=True)
```
